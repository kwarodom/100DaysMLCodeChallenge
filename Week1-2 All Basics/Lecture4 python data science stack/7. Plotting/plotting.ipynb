```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['price'] = boston['target']
```

```python
hdf = df.head(10)
hdf
```

```python
df.describe()
```

```python
hdf.style.highlight_max()
```

```python
hdf.style.background_gradient()
```

```python
hdf.style.bar()
```

```python
def odd_green(val):
    if val % 2 != 0:
        return 'color: green'
    return 'color:red'
```

```python
hdf.style.applymap(odd_green)
```

```python
def orange10(val):
    if val > 10:
        return 'background-color: orange'
    return ''
```

```python
hdf.style.applymap(odd_green).applymap(orange10)
```

```python
hdf.style.applymap(odd_green, subset=['ZN', 'RAD'])
```

```python
def top20(col):
    is_top20 = col > col.quantile(.8)
    return ['font-weight: bold' if v else '' for v in is_top20]
```

```python
hdf.style.apply(top20)
```

# customize pandas output

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-6, 6, 100)
ys = np.sinc(xs)
```

```python
plt.plot(xs, ys)
```

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.legend()
```

```python
plt.plot(xs, ys, label=r'$\frac{sin(x)}{x}$')
plt.plot(xs, np.sin(xs), label='sin(x)')
plt.legend()
```

```python
plt.subplot(2, 1, 1)
plt.plot(xs, np.sin(xs), label='sin')
plt.legend()

plt.subplot(2, 1, 2)
plt.fill(xs, -xs**2, label=r'${-x}^2$')
plt.legend()
```

```python
# plot  -6 < x < 6 
# 4 subplot (2,2)

plt.style.use('classic')

x2 = np.linspace(-6, 6, 100)

plt.subplot(2, 2, 1)
plt.plot(x2, np.sin(x2), label='sin')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, np.cos(x2), label='cos')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x2, np.tan(x2), label='tan')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x2, np.sinc(x2), label='sinc')
plt.legend()
```

```python
import pylab
import scipy
x = scipy.linspace(-2,2,1500)
y1 = scipy.sqrt(1-(abs(x)-1)**2)
y2 = -3*scipy.sqrt(1-(abs(x)/2)**0.5)
pylab.fill_between(x, y1, color='red')
pylab.fill_between(x, y2, color='red')
pylab.xlim([-2.5, 2.5])
```

```python
from ipywidgets import interact
```

```python
import numpy as np
```

```python
@interact(limit=6)
def plot_sin(limit):
    xs = np.linspace(-limit, limit, 100)
    plt.plot(xs, np.sin(xs), label='sin(x) [{} - {}]'.format(-limit, limit))
    plt.show()
```

```python
from bokeh.plotting import output_notebook
output_notebook()
```

```python
import pandas as pd
df = pd.read_csv('lnkd.csv', parse_dates=['Date'])
```

```python
from bokeh.plotting import figure, show
```

```python
fig = figure(x_axis_type='datetime')
fig.line(df['Date'], df['Close'])
show(fig)
```

```python
from bokeh.io import gridplot
```

```python
import numpy as np
```

```python
xs = np.linspace(-5, 5, 100)
fig1 = figure(title='sin(x)', plot_width=300, plot_height=300)
fig1.line(xs, np.sin(xs))
fig2 = figure(x_range=fig1.x_range, y_range=fig1.y_range, title='sinc(x)', plot_width=300, plot_height=300)
fig2.line(xs, np.sinc(xs))
gp = gridplot([[fig1, fig2]])
show(gp)
```
